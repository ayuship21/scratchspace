Here I imported the appropriate packages

```python
import pandas as pd
import matplotlib.pyplot as plt
```

I got the program to read the csv file so I could see which information they had on it.

```python
df = pd.read_csv('fish2.csv', encoding = 'unicode_escape')

df
```

I selected which columns I wanted from the data.

```python
desired_columns = [0, 5]
selected_data = df.iloc[:, desired_columns]
print(selected_data)
```

I constructed a bar chart based on the data. This served to analyze where fish were most commonly found. I will separately research temperature statistics in the regions with the most frequency of collected fish to analyze their preferred environment.



```python
fig, ax = plt.subplots()
item_counts = df['Site Name'].value_counts()

bar_width = 0.7

bar_color = 'blue'

ax.bar(item_counts.index, item_counts.values, width=bar_width, color=bar_color)

ax.set_xlabel('Site Name')
ax.set_ylabel('Frequency')
ax.set_title('Frequency of Fish Sites')

plt.xticks(rotation=90)

plt.xlim(item_counts.index[0], item_counts.index[-1120])

plt.show()
```
